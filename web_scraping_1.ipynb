```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.atpworldtour.com/en/rankings/singles?rankDate=2018-04-16&rankRange=1-1500'
response = requests.get(url)
```

```python
page = response.text
soup = BeautifulSoup(page,'html5lib')
```

```python
player_links = []
url_front = 'https://www.atpworldtour.com'
for element in soup.find_all(class_='player-cell'):
    player_links.append(url_front + element.find('a')['href'])
```

```python
years = ['2017','2016','2015','2014','2013','2012','2011','2010','2009']
player_stats_links = []
player_wins_links = []
for year in years:
    player_stats_links.extend([link.replace('overview','player-stats?year='+year+'&surfaceType=all') for link in player_links])
    player_wins_links.extend([link.replace('overview','player-activity?year='+year) for link in player_links])
```

Validation of length of player stats links length (100 players for 2017 to 2009 (9 years) is 900 stats links

```python
print(len(player_stats_links))
print(len(player_wins_links))
```

```python
players = []
skipped = 0
for i in range(len(player_stats_links)):
    response = requests.get(player_stats_links[i])
    if response.status_code == 200:
        soup2 = BeautifulSoup(response.text,'html5lib')
        stats = []
        name = soup2.find('div',id='pageHeader').find('h1').text.strip()
        year = soup2.find('div',class_='dropdown-label').text.strip()
        print(name,year)
        for element in soup2.find('div',id='playerMatchFactsContainer').find_all('td'):
            stats.append(element.text.strip())
        players.append(dict(zip(stats[::2],stats[1::2])))
        players[i-skipped]['Player'] = name
        players[i-skipped]['Year'] = year
        soup3 = BeautifulSoup(requests.get(player_wins_links[i]).text,'html5lib')
        if soup3.find(class_='players-stats-table'):
            winnings = soup3.find(class_='players-stats-table').find_all('td')[3].text.strip().split('\n',1)[0]
            record = soup3.find(class_='players-stats-table').find_all('td')[1].text.strip().split('\n',1)[0]
        else:
            winnings = 'None'
            record = 'None'
        players[i-skipped]['Winnings'] = winnings
        players[i-skipped]['Record'] = record
    else:
        skipped += 1
```
