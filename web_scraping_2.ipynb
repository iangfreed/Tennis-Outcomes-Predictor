### Grabbing Player Rankings 4/16/18

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.atpworldtour.com/en/rankings/singles?rankDate=2018-04-16&rankRange=1-1500'
response = requests.get(url)
page = response.text
soup = BeautifulSoup(page,'html5lib')
players = []
table_rows = soup.find('table').find('tbody').find_all('tr')
for row in table_rows:
    players.append({'Player': row.find_all('td')[3].get_text().strip(),'Ranking': row.find_all('td')[0].get_text().strip()})
```

```python
players
```

```python
import pickle

with open('players_rankings.pkl', 'wb') as picklefile:
    pickle.dump(players, picklefile)
```
